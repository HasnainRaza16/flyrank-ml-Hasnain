```python
import pandas as pd

df = pd.read_csv("flyrank-ml-hasnain/data/raw/content_refresh_anonymized.csv")

print("Total rows:", len(df))
print("Share declining (trend_direction == 'down'):", (df["trend_direction"] == "down").mean())
print("Avg CTR by decline status:")
print(df.groupby(df["trend_direction"] == "down")["ctr"].mean())
```

Out of 30,000 content items in the starter dataset, 54.2% are currently flagged as declining (trend_direction == "down"). Declining content shows a noticeably lower average CTR (0.32%) compared to non-declining content (0.73%)  less than half. This gap suggests CTR could be one of several signals worth investigating further, though at this stage it's just an observed association, not a confirmed cause of decline. With over half the dataset already declining, this lane looks worth pursuing for the next 7 weeks .There's clearly enough signal and volume to dig into.


## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

This analysis can say what signals are observed to correlate with content decline in this dataset..for example, declining content shows a lower average CTR than non-declining content. It can offer directional guidance, such as pointing toward CTR as a signal worth investigating further, and it can provide decision-support for a content strategist deciding where to focus review effort.
It cannot claim that low CTR causes decline, or that fixing CTR will reverse a decline that would require an experiment, not just an observed correlation. It also cannot claim anything about Google's actual ranking algorithm, since these are FlyRank's own observable measurements, not insight into search engine internals.
